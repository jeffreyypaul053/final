```python
from bs4 import BeautifulSoup
```

```python

import requests
```

```python
url="https://en.wikipedia.org/wiki/List_of_social_platforms_with_at_least_100_million_active_users"
```

```python

pro=requests.get(url)
```

```python
soup=BeautifulSoup(pro.text,"html")
print(soup)
```

```python
#heading
soup.find_all(['h1','h2','h3'])
```

```python
#paragraph
be=soup.find_all('p')
for pa in be:
    print(pa.get_text().strip())
```

```python
#table
tab=soup.find_all('table')[0]
print(tab)
```

```python
#table heading
col=tab.find_all('th')
col=[title.text.strip() for title in col]
print(col)
```

```python
import pandas as pd
```

```python

df=pd.DataFrame(columns=col)
```

```python
df
```

```python
web = tab.find_all('tr')
for scrap in web[1:101]: 
    dat = scrap.find_all('td')
    row = [data.text.strip() for data in dat]
    if not df.isin(row).all(axis=1).any():
        df.loc[len(df)] = row 
```

```python
df
```

```python
mov=df.to_csv(r'D:\Intership\task.csv',index=False)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
sn.set(color_codes=True)
```

```python
we=pd.read_csv('D:/Intership/task.csv')
```

```python

we.head()
```

```python

we.info()
```

```python
#visualisation
plt.figure(figsize=(20,18))
sn.scatterplot(x='Name',y='Monthly active users',data=we,s=100)
plt.title('Mostly used social media apps ',fontsize=20)
plt.xlabel('Name',fontsize=14)
plt.ylabel('Monthly active users',fontsize=14)
plt.xticks(rotation=90,fontsize=10)
plt.show()
```

```python
sn.barplot(x='Company', y='Country[b]', data=we,)
plt.title('Country users ', fontsize=16)
plt.xlabel('Company', fontsize=14)
plt.ylabel('Country[b] ', fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python
industry_counts = we['Company'].value_counts()
plt.figure(figsize=(20,18))
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', startangle=180, textprops={'fontsize': 14})
plt.title('Distribution of Companies by country', fontsize=20)
plt.axis('equal')  
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sn.boxplot(x='Company', y='Country[b]', data=we)
plt.title(' Revenue growth Distribution by Industry', fontsize=18)
plt.xlabel('Company', fontsize=16)
plt.ylabel('Country[b]', fontsize=16)
plt.xticks(rotation=90, fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,18))
sn.lineplot(x=' Company',y='Monthly active users',data=we,s=100)
plt.title('Mostly used social media apps ',fontsize=20)
plt.xlabel('Name',fontsize=14)
plt.ylabel('Monthly active users',fontsize=14)
plt.xticks(rotation=90,fontsize=10)
plt.show()
```
